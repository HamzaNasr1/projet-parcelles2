```python
import pandas as pd
import geopandas as gpd
import json
from fastapi import FastAPI
from fastapi.middleware.cors import CORSMiddleware

app = FastAPI()

# Autoriser les requêtes du front-end
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_methods=["*"],
    allow_headers=["*"],
)

# Charger les données
parcelles = gpd.read_file("DATA/parcelles.shp")

# Convertir les colonnes datetime en string pour éviter l'erreur
for col in parcelles.columns:
    if pd.api.types.is_datetime64_any_dtype(parcelles[col]):
        parcelles[col] = parcelles[col].astype(str)

# Convertir en GeoJSON
parcelles_geojson = json.loads(parcelles.to_json())

@app.get("/parcelles")
def get_parcelles():
    return parcelles_geojson
```
